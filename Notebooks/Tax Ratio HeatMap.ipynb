```python
#Import libraries

import pandas as pd
import folium
from folium.plugins import HeatMap
import branca.colormap
from collections import defaultdict
```

```python
#Read in CSV files and initialize Dataframe

ACTaxes = pd.read_csv(r'C:\Users\Noah Hellenthal\Desktop\Real Estate\Tables\AkronCantonTaxes.csv')
```

```python
#Initialize map data

max_amount = float(ACTaxes['2018_Tax'].max())
print(max_amount)
lon_avg = ACTaxes['Lon'].mean()
print(lon_avg)
lat_avg = ACTaxes['Lat'].mean()
print(lat_avg)
```

```python
#Create heat map
gradient = {0.001: 'yellow', 0.025: 'orange', 0.05: 'red'}
hmap = folium.Map(location=[lat_avg, lon_avg], zoom_start=10)
hm_wide = HeatMap( list(zip(ACTaxes.Lat.values, ACTaxes.Lon.values, ACTaxes['TaxRatio'].values)),
                 min_opacity=0.2,
                 max_val=max_amount,
                 radius=10,
                 gradient = gradient,
                 blur=15,
                 max_zoom=10)
hmap.add_child(hm_wide)

#YlOrRd_09
colormap = branca.colormap.linear.YlOrRd_09.scale(0,0.05).to_step(10)
colormap.caption = 'Tax Ratio'
hmap.add_child(colormap) #Taxes divided by price of the home would logically yield a lower ratio value
#All values are within (0, 0.03)
#Ratio of Tax to calculated monthly payment? <- Problems: Term, Rate would have to be assumed
#Display only most recent tax value?
#Heatmaps may be showing magnitude instead of actual value
```
