```python
df = pd.read_csv("AkronCantonTaxes.csv")

origin = [40.966186, -81.448951]
```

```python
lat = df.Lat
lon = df.Lon

loc = pd.concat([lat, lon], axis = 1)
print(loc)
locationlist = loc.values.tolist()
len(locationlist)
locationlist[1]
```

```python
#Create maps to show points where houses are for sale

map = folium.Map(location=origin, zoom_start=10)
homes = plugins.MarkerCluster().add_to(map)

for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point]).add_to(homes)
map
```

```python
df = pd.DataFrame()
df['Latitude'] = lat
df['Longitude'] = lon

heat_df = [[row['Latitude'], row['Longitude']] for index, row in df.iterrows()]

map_heat = folium.Map(location = origin, zoom_start=10)
HeatMap(heat_df).add_to(map_heat)
map_heat
```
