```python
#Import libraries

import pandas as pd
import folium
from folium.plugins import HeatMap
import branca.colormap
from collections import defaultdict
```

```python
#Read in CSV files and initialize Dataframe

ACTaxes = pd.read_csv(r'C:\Users\Noah Hellenthal\Desktop\Real Estate\AkronCantonTaxes.csv')
```

```python
#Initialize map data

max_amount = float(ACTaxes['2018_Tax'].max())
print(max_amount)
lon_avg = ACTaxes['Lon'].mean()
print(lon_avg)
lat_avg = ACTaxes['Lat'].mean()
print(lat_avg)
```

```python
#Create heat map
gradient = {0.2: 'yellow', 0.7: 'orange', 1: 'red'}
hmap = folium.Map(location=[lat_avg, lon_avg], zoom_start=10)
hm_wide = HeatMap( list(zip(ACTaxes.Lat.values, ACTaxes.Lon.values, ACTaxes['2018_Tax'].values)),
                 min_opacity=0.2,
                 max_val=max_amount,
                 radius=10,
                 gradient = gradient,
                 blur=15,
                 max_zoom=10)
hmap.add_child(hm_wide)

#YlOrRd_09
colormap = branca.colormap.linear.YlOrRd_09.scale(0,max_amount).to_step(10)
colormap.caption = '2018 Taxes'
hmap.add_child(colormap)
```

```python
maptax = folium.Map(location=[lat_avg, lon_avg], zoom_start=10)

steps = 20
colormap = branca.colormap.linear.YlOrRd_09.scale(0,1).to_step(steps)
gradient_map = defaultdict(dict)
for i in range(steps):
    gradient_map[1/steps*i] = colormap.rgb_hex_str(1/steps*i)
colormap.add_to(maptax)

HeatMap(list(zip(ACTaxes.Lat.values, ACTaxes.Lon.values, ACTaxes['2018_Tax'].values)),
       gradient = gradient_map).add_to(maptax)
maptax
```
